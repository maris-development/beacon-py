```python
from client import Client
from sqlalchemy import *
```

```python
client = Client("http://localhost:5001")
client.list_tables()
```

```python
emodnet_chemistry = client.table('emodnet_chemistry')
```

```python
query = query = select(
    emodnet_chemistry.longitude,
    emodnet_chemistry.latitude,
    emodnet_chemistry.depth
).where(
    and_(
        emodnet_chemistry.latitude > 40,
        emodnet_chemistry.latitude < 60,
        emodnet_chemistry.longitude > -20,
        emodnet_chemistry.longitude < 0,
        emodnet_chemistry.depth < 10
    )
).limit(1000000)

client.query_to_pandas(query)
```
